# SOLUTIONS: Can we predict which emerging compounds will have a high extent of effect (EoE) across Europe?


```python
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pylab
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
import statsmodels.api as sm
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn import svm
from sklearn.metrics import mean_squared_error

sns.set()
```

```python
def evaluate_test(Y, Y_h, dat):
    colors = {'1_Pharma':'red', '2_Pest':'blue', '3_REACH':'black'}
    spacing = {'1_Pharma':1, '2_Pest':2, '3_REACH':3}

    my = [10e-6, 10e3]
    mx = [10e-6, 10e3]
    RMSE = np.sqrt(mean_squared_error(Y, Y_h))
    # plot those in the test set
    plt.scatter(np.exp(Y), np.exp(Y_h), c = dat['chem'].loc[Y.index].apply(lambda x: colors[x]), label = 'RMSE = %.2f' % RMSE)

    #c=df['color'].apply(lambda x: colors[x])
    plt.plot(mx, my, '--k')
    #plt.colorbar()
    plt.xscale('log')
    plt.yscale('log')
    plt.xlim(mx)
    plt.ylim(my)
    plt.title('test set')
    plt.xlabel('Observed logEoE')
    plt.ylabel('Predicted logEoE')
    plt.legend()

    fig = plt.figure()
    ax = Axes3D(fig)
    im = ax.scatter(Y, dat['chem'].loc[Y.index].apply(lambda x: spacing[x]), Y_h,  c = dat['chem'].loc[Y.index].apply(lambda x: colors[x]), cmap = 'jet')
    ax.set_xlim3d(-8, 4)
    ax.set_ylim3d(1,3)
    ax.set_zlim3d(-8,4)                                                    

def plot_residuals(Y, Y_h, dat):
    spacing = {'1_Pharma':1, '2_Pest':2, '3_REACH':3}
    for ii, cc in enumerate(spacing.keys()):
        # find indices of chemicals for this class
        ind = dat[dat['chem'] == cc].index
        # find the subset of these indices that were in the test set
        sub_ind = Y.index.intersection(ind)
        # find the row number for those rows whose indicies reside in the indicies corresponding to the class
        loc = [ii for ii,jj in enumerate(Y) if Y.index[ii] in sub_ind]
        diff = Y.loc[sub_ind] - Y_h[loc]
        plt.figure(1)
        plt.subplot('3' + '1' + str(ii))
        plt.hist(diff, bins = 20)
        plt.xlim([-3, 3])
        plt.title(cc + ' residuals')
        plt.figure(2)
        plt.subplot('3' + '1' + str(ii))
        sns.violinplot(diff)
        plt.xlim([-4, 4])
        plt.title(cc + ' residuals')
    plt.figure(1)
    plt.tight_layout()
    plt.figure(2)
    plt.tight_layout()
```

# Import the data 
The data we use comes from the inputs to and outputs from the SOUTIONS modelling train. The inputs include chemical properties and use volumes. The main output is the extent of effect (EoE), a statistical property which represents the 95th spatial percentile of the 99th temporal percentile of substance concentration over toxicity.

```python
dat = pd.read_excel('../data/ResultsCaseStudies_EoE_FoE.xlsx', sheetname = 'EoE_FoE')
```

```python
dat.columns
```

# Examine and clean the data

```python
dat.isnull().sum()
```

```python
dat.plot.box()
plt.xticks(rotation = '90'); 
plt.yscale('log')
```

We clean the dataset by removing columns with poor data coverage, data with redundant/colinear information, and data that is not useful for the model. <br>
We also need to one-hot the data for the chemical class.

```python
len_before = len(dat)
dat[dat['E/PNEC'] > 1e7] = np.nan
dat[dat['E2RIVTOT (kg/y)'] < 0] = np.nan
dat[dat['EoE'] < 10e-5] = np.nan

dat.drop(['case','logPAF'], axis = 1, inplace = True)
dat.dropna(axis = 0, inplace = True)
chemical = pd.get_dummies(dat['chem'])
#dat.drop(['chem'], axis = 1, inplace = True)
dat = pd.concat([dat, chemical], axis = 1)
dat.dropna(inplace = True, axis = 0)
dat.columns
len_after = len(dat)
```

```python
print(len_before, len_after)
dat
```

Select the features for use in the machine learning algorithm and scale to the logarithm for values with a multi-order-of-magnitude range

```python
X = dat[['1_Pharma','2_Pest','3_REACH','logE/PNEC','E2RIVTOT (kg/y)','Etot (kg/y)','kdw','logKOW0','MW', 'EoE', 'chem']].copy()
X['E2RIVTOT (kg/y)'] = np.log(X['E2RIVTOT (kg/y)'])
X['Etot (kg/y)'] = np.log(X['Etot (kg/y)'])
X['kdw'] = X['kdw'] * -1
X['EoE'] = np.log(X['EoE'])
X.plot.box()
plt.xticks(rotation = '90'); 
plt.yscale('log')
```

Pairplot of features including target colored by chemical class

```python
sns.pairplot(X.drop(['1_Pharma','2_Pest','3_REACH'], axis = 1), hue = 'chem')
```

Assign target and features

```python
X.drop('chem', axis = 1, inplace = True)
Y = X.pop('EoE')
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.8, random_state=52354)
```

# Build Random Forest model

```python
for depth in range(6,10):
    rf = RandomForestRegressor(n_estimators=5000, max_depth = depth, random_state = 42)
    rf.fit(X_train, Y_train)
    feature_importance_data = []
    features = X_train.columns
    for feature_name, feature_importance in zip(features, rf.feature_importances_):
        feature_importance_data.append({'feature' : feature_name, 'importance' : feature_importance})
    
    feature_importance_data = pd.DataFrame(feature_importance_data)
    plt.figure(5)
    plt.subplot('2' + '2' + str(depth-5))
    Y_h = rf.predict(X_train)
    plt.scatter(np.exp(Y_train), np.exp(Y_h))
    plt.title('depth = ' + str(depth))
    my = [10e-6, 10e3]
    mx = [10e-6, 10e3]
    plt.plot(mx, my, '--k')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlim(mx)
    plt.ylim(my)
    plt.xlabel('O logEoE')
    plt.ylabel('P logEoE')
    plt.tight_layout()
```

```python
feature_importance_data
```

```python
feature_importance_data.set_index('feature', inplace = True)
feature_importance = feature_importance_data.sort_values(by = 'importance', ascending = False)
ax = feature_importance.plot(kind = 'barh', y = 0)
```

```python
Y_h = rf.predict(X_test)
evaluate_test(Y_test, Y_h, dat)
```

Examine the absolute error per substance class

```python
plot_residuals(Y_test, Y_h, dat)
```

```python
Y_h = clf.predict(X_test)
evaluate_test(Y_test, Y_h, dat)
```
